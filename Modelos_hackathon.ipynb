```python
import pandas as pd
import numpy as np
import datetime
now = datetime.datetime.now()
from sklearn.neural_network import MLPRegressor # ANN
from sklearn.neighbors import KNeighborsRegressor # KNN Regressor
from sklearn.svm import SVR # Support Vetor Regressor #
from sklearn.gaussian_process import GaussianProcessRegressor 
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.linear_model import SGDRegressor
# Pre procesamiento
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# Serializacion de modelos
import pickle
```

```python
df = pd.read_csv("/Users/carlo/OneDrive/Documentos/DataScience/portafolio/Hackathon/data_hackathon_v4.csv", sep=";")
df = df.set_index('sample_id')
df[['plantacion']] = now.year - df[['plantacion']]
# One Hot Encoding
encode = ['variedad','valle','sistema_conduccion']
for col in encode:
    dummy = pd.get_dummies(df[col],prefix=col)
    df = pd.concat([df,dummy],axis=1)
    del df[col]
```

```python
# Generación de los 7 df para los 70 outputs
# Cantidad de semillas
df_cantidad_semillas = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)', 
                        'cantidad_semillas']]
# Separación de las features y el target 
cantidad_semillas_X = df_cantidad_semillas.drop('cantidad_semillas', axis = 1)
cantidad_semillas_y = df_cantidad_semillas['cantidad_semillas']
# Cantidad de semillas - separación entre entrenamiento y testeos
X_train_o1, X_test_o1, y_train_o1, y_test_o1 = train_test_split(cantidad_semillas_X, cantidad_semillas_y, test_size=0.2, random_state=123)


# Cantidad de hollejo
df_cantidad_hollejo = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',                       'cantidad_hollejo']]
# Separación de las features y el target 
cantidad_hollejo_X = df_cantidad_hollejo.drop('cantidad_hollejo', axis = 1)
cantidad_hollejo_y = df_cantidad_hollejo['cantidad_hollejo']
# Cantidad de semillas - separación entre entrenamiento y testeos
X_train_o2, X_test_o2, y_train_o2, y_test_o2 = train_test_split(cantidad_hollejo_X, cantidad_hollejo_y, test_size=0.2, random_state=123)

# Humedad del orujo
df_humedad_orujo = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',                    'humedad_orujo']]
# Separación de las features y el target 
humedad_orujo_X = df_humedad_orujo.drop('humedad_orujo', axis = 1)
humedad_orujo_y = df_humedad_orujo['humedad_orujo']
# Humedad de orujo - separación entrenamiento y testeo
X_train_o3, X_test_o3, y_train_o3, y_test_o3 = train_test_split(humedad_orujo_X, humedad_orujo_y, test_size=0.2, random_state=123)

# Cantidad de polifenoles totales
df_polifenoles_totales = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',
                          'polifenoles_totales']]
# Separación de las features y el target 
polifenoles_totales_X = df_polifenoles_totales.drop('polifenoles_totales', axis = 1)
polifenoles_totales_y = df_polifenoles_totales['polifenoles_totales']
# Cantidad de polifenoloes - Separacion entrenamiento y testeo
X_train_o4, X_test_o4, y_train_o4, y_test_o4 = train_test_split(polifenoles_totales_X, polifenoles_totales_y, test_size=0.2, random_state=123)


# Cantidad de taninos
df_taninos = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',              'taninos' ]]
# Separación de las features y el target 
taninos_X = df_taninos.drop('taninos', axis = 1)
taninos_y = df_taninos['taninos']
# Taninos - Separacion entrenamiento y testeo
X_train_o5, X_test_o5, y_train_o5, y_test_o5 = train_test_split(taninos_X, taninos_y, test_size=0.2, random_state=123)


# Cantidad de flavanoles
df_flavanoles = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',
                 'flavanoles']]
# Separación de las features y el target 
flavanoles_X = df_flavanoles.drop('flavanoles', axis = 1)
flavanoles_y = df_flavanoles['flavanoles']
# Flavanoles - Separacion entrenamiento y testeo
X_train_o6, X_test_o6, y_train_o6, y_test_o6 = train_test_split(flavanoles_X, flavanoles_y, test_size=0.2, random_state=123)

# Cantidad de acidos fenolicos, sumatoria entre acidos hidroxicinamicos y acidos hidrocibenzoicos
df_acidos_fenolicos = df[['rendimiento', 'riego', 'DGA', 'plantacion', 'cosecha', 'nitrogeno','maceracion', 'temperatura_maceracion', 'variedad_Cabernet Sauvignon','variedad_Carmenere', 'variedad_Cinsault', 'variedad_Malbec',       'variedad_Merlot', 'variedad_Pinot Noir', 'variedad_Syrah','valle_Cachapoal', 'valle_Casablanca', 'valle_Colchagua', 'valle_Itata','valle_Maipo  ', 'valle_Maipo alto', 'valle_Maule','sistema_conduccion_Parron (3x3)', 'sistema_conduccion_espaldera (1x2)','sistema_conduccion_espaldera (1x3)',
                   'acidos_fenolicos']]
# Separación de las features y el target 
acidos_fenolicos_X = df_acidos_fenolicos.drop('acidos_fenolicos', axis = 1)
acidos_fenolicos_y = df_acidos_fenolicos['acidos_fenolicos']
# Cantidad de acidos fenolicos - Separación entrenamiento y testeo
X_train_o7, X_test_o7, y_train_o7, y_test_o7 = train_test_split(flavanoles_X, flavanoles_y, test_size=0.2, random_state=123)
```

```python
names = ["Nearest_Neighbors", 
         "Linear_SVM", 
         "Polynomial_SVM", 
         "RBF_SVM", 
         "Neural_Net",
         "Gaussian_Process",
         "Decision_Tree"]
regressors = [KNeighborsRegressor(3),
              SVR(kernel="linear", C=0.025),
              SVR(kernel="poly", degree=3, C=0.025),
              SVR(kernel="rbf", C=1, gamma=2),
              MLPRegressor(alpha=1, max_iter=1000),
              GaussianProcessRegressor(),
              DecisionTreeRegressor(max_depth=5)]
standar_scaler = StandardScaler()
```

# Entrenamiento modelos para cantidad de semillas

```python
# Se genera un arreglo vacio 
scores = []
# En un ciclo FOR se hace el entrenamiento de los clasificadores 
for name, clf in zip(names, regressors):
    clf.fit(X_train_o1,  y_train_o1)
    score = clf.score(X_test_o1, y_test_o1)
    scores.append(score)
# Guardar los resultado en un dataframe para reunirlos todos 
df_o1 = pd.DataFrame()
df_o1['name'] = names
df_o1['scores_o1'] = scores
df_o1
```
